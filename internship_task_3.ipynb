```python
import pandas as pd

data = pd.DataFrame({
    "stage": ["Visitors", "Leads", "Qualified Leads", "Opportunities", "Customers"],
    "count": [5000, 1200, 600, 300, 120]
})

data
```

```python
data['conversion_rate'] = data['count'] / data['count'].shift(1)
data
```

```python
data['conversion_rate'] = data['conversion_rate'].fillna(1)
data
```

```python
data['drop_off_%'] = (1 - data['conversion_rate']) * 100
data
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(data['stage'], data['count'])
plt.title("Marketing Funnel")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure()
plt.plot(data['stage'], data['conversion_rate'], marker='o')
plt.title("Conversion Rate by Stage")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure()
plt.bar(data['stage'], data['drop_off_%'])
plt.title("Drop-off Percentage")
plt.xticks(rotation=30)
plt.show()
```

**Insights:**


*  Large drop-off from Visitors to Leads


*   Conversion decreases at each stage
*  Biggest loss happens in early funnel stages


*  Final conversion to customers is low compared to initial traffic






**Recommendations:**


*   Improve landing page to convert more visitors
*   Use better targeting for quality leads


*  Follow up faster with leads
*  Optimize marketing campaigns
